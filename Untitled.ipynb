```python
# Lead Scoring for Data Science Training & Admissions

# 1. Import Libraries

!pip install xgboost
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# 2. Load Dataset
data = pd.read_csv("Lead_Scoring_Enquiries_4000.csv")
print("Data Shape:", data.shape)
print(data.head())

# 3. EDA
print(data.info())
print(data.describe())

# Missing values
plt.figure(figsize=(10,5))
sns.heatmap(data.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()

# Target distribution
sns.countplot(x='Converted', data=data)
plt.title("Target Variable Distribution")
plt.show()

# Categorical feature distribution
plt.figure(figsize=(8,5))
data['Email_Source'].value_counts().plot(kind='bar')
plt.title("Leads by Email Source")
plt.show()

# 4. Data Preprocessing
# Fill missing values
data.fillna({'Profession':'Unknown', 'Location':'Unknown'}, inplace=True)
data['Engagement_Score'].fillna(data['Engagement_Score'].median(), inplace=True)

# Encode categorical variables
le = LabelEncoder()
for col in ['Email_Source','Contacted','Location','Profession','Course_Interest']:
    data[col] = le.fit_transform(data[col])

# Features & Target
X = data.drop(columns=['Lead_ID','Name','Converted'])
y = data['Converted']

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# 5. Model Training
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]
    print("\n====", name, "====")
    print(classification_report(y_test, y_pred))
    print("ROC-AUC:", roc_auc_score(y_test, y_proba))

# 6. Lead Scoring & Ranking
best_model = models['XGBoost']  # choose best performing
lead_scores = best_model.predict_proba(X_scaled)[:,1]
data['Lead_Score'] = lead_scores

# Rank leads
ranked_leads = data.sort_values(by='Lead_Score', ascending=False)
ranked_leads[['Lead_ID','Name','Lead_Score','Converted']].head(20)

# Save results
ranked_leads.to_csv("Lead_Scoring_Output.csv", index=False)

# 7. Visualization of Insights
# ROC Curve
fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(X_test)[:,1])
plt.plot(fpr, tpr, label='XGBoost')
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# Conversion by Source
plt.figure(figsize=(8,5))
sns.barplot(x='Email_Source', y='Converted', data=data)
plt.title("Conversion Rate by Source")
plt.show()

# 8. Simple Prediction Function
def predict_new_lead(new_data):
    df = pd.DataFrame([new_data])
    for col in ['Email_Source','Contacted','Location','Profession','Course_Interest']:
        df[col] = le.fit_transform(df[col])
    df_scaled = scaler.transform(df)
    score = best_model.predict_proba(df_scaled)[:,1][0]
    return score

# Example prediction
example = {
    'Email_Source': 'Website',
    'Contacted': 'Yes',
    'Engagement_Score': 85,
    'Location': 'Mumbai',
    'Profession': 'Student',
    'Course_Interest': 'Data Science – Beginner',
    'Follow_Up_Count': 2
}
print("Predicted Conversion Probability:", predict_new_lead(example))
```

```python
# Lead Scoring for Data Science Training & Admissions

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Try importing XGBoost safely
try:
    from xgboost import XGBClassifier
    xgb_available = True
except ImportError:
    print("⚠️ XGBoost not installed. Skipping XGBoost model.")
    xgb_available = False

# 2. Load Dataset
data = pd.read_csv("Lead_Scoring_Enquiries_4000.csv")
print("Data Shape:", data.shape)
print(data.head())

# 3. EDA
print(data.info())
print(data.describe())

# Missing values
plt.figure(figsize=(10,5))
sns.heatmap(data.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()

# Target distribution
sns.countplot(x='Converted', data=data)
plt.title("Target Variable Distribution")
plt.show()

# Categorical feature distribution
plt.figure(figsize=(8,5))
data['Email_Source'].value_counts().plot(kind='bar')
plt.title("Leads by Email Source")
plt.show()

# 4. Data Preprocessing
# Fill missing values
data.fillna({'Profession':'Unknown', 'Location':'Unknown'}, inplace=True)
data['Engagement_Score'].fillna(data['Engagement_Score'].median(), inplace=True)

# Encode categorical variables
le = LabelEncoder()
for col in ['Email_Source','Contacted','Location','Profession','Course_Interest']:
    data[col] = le.fit_transform(data[col])

# Features & Target
X = data.drop(columns=['Lead_ID','Name','Converted'])
y = data['Converted']

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# 5. Model Training
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)
}

if xgb_available:
    models['XGBoost'] = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]
    print("\n====", name, "====")
    print(classification_report(y_test, y_pred))
    print("ROC-AUC:", roc_auc_score(y_test, y_proba))

# 6. Lead Scoring & Ranking
best_model = models['Random Forest'] if not xgb_available else models['XGBoost']
lead_scores = best_model.predict_proba(X_scaled)[:,1]
data['Lead_Score'] = lead_scores

# Rank leads
ranked_leads = data.sort_values(by='Lead_Score', ascending=False)
ranked_leads[['Lead_ID','Name','Lead_Score','Converted']].head(20)

# Save results
ranked_leads.to_csv("Lead_Scoring_Output.csv", index=False)

# 7. Visualization of Insights
# ROC Curve
fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(X_test)[:,1])
plt.plot(fpr, tpr, label='Best Model')
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# Conversion by Source
plt.figure(figsize=(8,5))
sns.barplot(x='Email_Source', y='Converted', data=data)
plt.title("Conversion Rate by Source")
plt.show()

# 8. Simple Prediction Function
def predict_new_lead(new_data):
    df = pd.DataFrame([new_data])
    for col in ['Email_Source','Contacted','Location','Profession','Course_Interest']:
        df[col] = le.fit_transform(df[col])
    df_scaled = scaler.transform(df)
    score = best_model.predict_proba(df_scaled)[:,1][0]
    return score

# Example prediction
example = {
    'Email_Source': 'Website',
    'Contacted': 'Yes',
    'Engagement_Score': 85,
    'Location': 'Mumbai',
    'Profession': 'Student',
    'Course_Interest': 'Data Science – Beginner',
    'Follow_Up_Count': 2
}
print("Predicted Conversion Probability:", predict_new_lead(example))

# 9. Streamlit App Code (save this as app.py to run: `streamlit run app.py`)

app_code = """
import streamlit as st
import pandas as pd

st.title("🎯 Lead Scoring Prediction App")

st.write("Upload a CSV file of new leads to predict conversion probability")

uploaded_file = st.file_uploader("Choose a CSV file", type=["csv"])

if uploaded_file is not None:
    new_leads = pd.read_csv(uploaded_file)
    st.write("Uploaded Data:")
    st.dataframe(new_leads.head())

    # Encode and scale (reuse preprocessing pipeline)
    for col in ['Email_Source','Contacted','Location','Profession','Course_Interest']:
        new_leads[col] = le.fit_transform(new_leads[col])
    new_scaled = scaler.transform(new_leads)

    # Predict scores
    scores = best_model.predict_proba(new_scaled)[:,1]
    new_leads['Lead_Score'] = scores
    new_leads = new_leads.sort_values(by='Lead_Score', ascending=False)

    st.write("### Predicted Lead Scores")
    st.dataframe(new_leads[['Lead_Score'] + [col for col in new_leads.columns if col != 'Lead_Score']])

    # Download option
    csv = new_leads.to_csv(index=False).encode('utf-8')
    st.download_button("Download Results", csv, "Lead_Scoring_Results.csv", "text/csv")
"""

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("✅ Streamlit app code saved as app.py. Run with: streamlit run app.py")
```

```python
import streamlit as st
import pandas as pd
import pickle

# Load trained model & preprocessing
with open("conversion_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("conversion_preprocess.pkl", "rb") as f:
    le, scaler = pickle.load(f)

st.title("Lead Scoring Prediction App")

uploaded_file = st.file_uploader("Upload CSV file of leads", type=["csv"])

if uploaded_file is not None:
    new_leads = pd.read_csv(uploaded_file)

    st.subheader("Uploaded Data")
    st.dataframe(new_leads.head())

    # Encode categorical columns
    for col in ['Email_Source','Contacted','Location','Profession','Course_Interest']:
        new_leads[col] = le.fit_transform(new_leads[col])

    # Scale numeric data
    new_scaled = scaler.transform(new_leads)

    # Predict conversion probabilities
    scores = model.predict_proba(new_scaled)[:,1]
    new_leads['Lead_Score'] = scores

    # Show results
    st.subheader("Predicted Lead Scores")
    st.dataframe(new_leads.sort_values(by='Lead_Score', ascending=False))

    # Download option
    csv = new_leads.to_csv(index=False).encode("utf-8")
    st.download_button("Download Results", csv, "Lead_Scoring_Results.csv", "text/csv")
```
